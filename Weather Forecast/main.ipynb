```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px
```

```python
data = pd.read_csv('DailyDelhiClimateTrain.csv')
data.head()
```

```python
data.describe()
```

```python
figure = px.line(data, x='date', y='meantemp', title='Mean Temperature in Delhi Over the Years')
figure.show()
```

```python
figure = px.line(data, x="date", 
                 y="humidity", 
                 title='Humidity in Delhi Over the Years')
figure.show()
```

```python
figure = px.line(data, x="date", 
                 y="wind_speed", 
                 title='Wind Speed in Delhi Over the Years')
figure.show()
```

```python
figure = px.scatter(data_frame = data, x="humidity",
                    y="meantemp", size="meantemp", 
                    trendline="ols", 
                    title = "Relationship Between Temperature and Humidity")
figure.show()
```

# Extract data from date column into year and month

```python
data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d')
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month

data.head()
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15, 10))
plt.title("Temperature Change in Delhi Over the Years")
sns.lineplot(data = data, x='month', y='meantemp', hue='year')
plt.show()
```

# Create Forecast

```python
forecast_data = data.rename(columns={'date' : 'ds', 'meantemp' : 'y'})

forecast_data.head()
```

```python
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly

model = Prophet()
model.fit(forecast_data)
```

```python
forecast = model.make_future_dataframe(periods=365)
prediction = model.predict(forecast)
```

```python
plot_plotly(model, prediction)
```
